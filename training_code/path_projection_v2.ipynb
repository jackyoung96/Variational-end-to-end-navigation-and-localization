```python
from PIL import Image, ImageDraw
import numpy as np
import matplotlib.pyplot as plt
import os
import math
import aggdraw
```

```python
path = 'data/'

folder_list = os.listdir(path)

for folder in folder_list:
    imgs = os.listdir(path+folder)
    for img in imgs:
        if img.split('_')[0]=='center':
            print(path+folder+'/'+img)
```

```python
path = 'data/892/center_0.0295.png'
image_sample = Image.open(path)
image_arr = np.asarray(image_sample)
plt.imshow(image_arr)
plt.show()
```

```python
theta = float(path.split('_')[1].split('.png')[0])
# R=1/(math.pi*theta/180)
R = -1/theta

print(R)
h=1.7
h_t = 1280/720*h
y0=1.6

p=360

u0 = 540
v0 = 360

num = 1000

y = np.linspace(h,min(abs(R-y0)*2/3,50),num)
if R >= 0 :
    x = R - np.sqrt(R**2 - (y+y0)**2)
else :
    x = R + np.sqrt(R**2 - (y+y0)**2)  
    
u = (p*x*h_t/y/h + u0).astype(np.int32)
v = (p*h_t/y + v0).astype(np.int32)


point_list = []
for i in range(num):
    point_list.append((u[i],v[i]))

im_path = ImageDraw.Draw(image_sample)
im_path.point(point_list, fill= 'red')
image_arr = np.asarray(image_sample)
plt.imshow(image_arr)
```
